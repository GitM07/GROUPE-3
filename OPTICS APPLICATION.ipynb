```python
import numpy as np
import pandas as pd
from sklearn.cluster import OPTICS
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("/Mall_Customers.csv", sep=',')
```

```python
data
```

```python
data = data.drop(['CustomerID','Gender'],axis=1)
```

```python
x = data.iloc[:, 0]
y = data.iloc[:, 1]
z = data.iloc[:, 2]

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='m', marker='o')

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')

plt.show()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
columns = data.columns
newdata = pd.DataFrame(scaled_data, columns=columns)
```

```python
newdata.head()
```

```python
x = newdata.iloc[:, 0]
y = newdata.iloc[:, 1]
z = newdata.iloc[:, 2]
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')

plt.show()
```

```python
optics =OPTICS(min_samples=10)
optics.fit(newdata)
```

```python
labels = optics.labels_
```

```python
np.unique(labels)
```

```python
reachability = optics.reachability_

plt.bar(range(len(newdata)), reachability)
plt.xlabel("Data point index")
plt.ylabel("Reachability distance")
plt.title("Reachability plot")

threshold = 0.5
plt.axhline(y=threshold, color="red", linestyle="--")

plt.show()
```

```python
n_clusters = len(np.unique(labels))
print("Number of clusters:", n_clusters)
```

```python
data['groups'] = np.nan
for i in range(len(data)):
    data.loc[i, 'groups'] = labels[i]
```
